```python
import numpy as np
```

```python
# Array of Zeros

arr_zeros= np.zeros((4,4))
print(arr_zeros)
print()

# Empty Arrray
arr_ones= np.ones((4,4))
print(arr_ones)
print()

# Array of Ones
arr_empty= np.empty((4,4))
print(arr_empty)
print()


```

```python
arr_arrange = np.arange(100,10000,50)
print(arr_arrange)
print()
```

```python
# Array Reshape
l =arr_arrange.size
r = 6
c = l//r
arr_reshaped = arr_arrange.reshape((r,c))
print(arr_reshaped)
print()
```

```python
arr_linspace = np.linspace(20,1000)
print(arr_linspace)
print()
```

```python
sl=arr_linspace.size
row = 5
col = sl//row

final_arr = arr_linspace.reshape((row,col))
print(final_arr)
print()
```
